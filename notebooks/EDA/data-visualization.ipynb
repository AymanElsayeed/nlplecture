# Data Visualization

Data sets:
   - iris
   - titanic
   - tips


Packages:
   - Seaborn
   - Matplotlib

Plots:

- Bar plot
- Violin plot
- Scatter plot
- Count plot
- Heat map
- Line plot
- Histogram
- Pie Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
tips = sns.load_dataset("tips")
```

## Histogram

Shows the distribution of numerical variable

- Centra Tendency
- Skewness

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=tips, x='total_bill', bins=30)
plt.title('Distribution of Total Bills')
plt.show()
```

```python
tips["total_bill"].describe()
```

# Line Plot

Shows change in the data over "time"

- cycical pattern
- trends
- compare

```python

plt.figure(figsize=(10, 6))
daily_avg = tips.groupby('day')['total_bill'].mean()
sns.lineplot(data=daily_avg)
plt.title('Average Bill Amount by Day')
plt.show()
```

# Scatter Plot

shows the relationship between two variables

- positive correlation
- negative correlation
- linear correlation
- no correlation

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Width by Species')
plt.show()
```

# Bar Plot

Shows distribution of categorical variable

Good to compare two categorical variables

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=tips, x='day', y='total_bill')
plt.title('Average Total Bill by Day')
plt.show()
```

# Box Plot

compare distributions of numerical variables

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.title('Distribution of Sepal Length by Species')
plt.show()
```

# Other Plots

## Violin Plot

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=iris, x='species', y='sepal_length')
plt.title('Distribution of Sepal Length by Species')
plt.show()
```

## Count Plot

```python

plt.figure(figsize=(10, 6))
sns.countplot(data=titanic, x='class')
plt.title('Number of Passengers by Class')
plt.show()
```

## Pie Plot

```python

plt.figure(figsize=(8, 8))
survival_counts = titanic['survived'].value_counts()
plt.pie(survival_counts, labels=['Did Not Survive', 'Survived'],
        autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'])
plt.title('Survival Distribution on Titanic')
plt.show()
```
